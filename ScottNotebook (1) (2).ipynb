# **Playgrounds.csv**

Introduction: 
We will be determing which neighborhood is the best for outdoor activities. 
The Metric: 
The metric we will be using to determine this is the number of playgrounds in each neighborhood.  

```python
import pandas as pd
import numpy as np
import geopandas
import csv
import collections
import matplotlib.pyplot as plt

#shp is shapefile
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
```

```python
playgrounds = collections.Counter()
with open('Playgrounds.csv') as file:
    for x in csv.reader(file, delimiter=','):
        playgrounds[x[7]] += 1

df3 = pd.DataFrame(playgrounds.most_common())
df3.columns=['Neighborhood', 'Number of Playgrounds']
df3.groupby("Neighborhood").sum()['Number of Playgrounds']

shaped3 = neighborhoods.merge(df3, how='left', left_on='hood', right_on='Neighborhood')
shaped3.plot(column='Number of Playgrounds',
                    cmap='Purples',
                    edgecolor="black",
                    legend=True,
                    legend_kwds={'label':"Number of Playgrounds"},
                    figsize=(15,10),
                    missing_kwds={"color":"lightgrey"})

# .head() gets top values from DataFrame
topValues = df3.head(15)
print(topValues)
ax = topValues.plot(kind="bar", figsize=(10,6), colormap='terrain')
ax.set_xticklabels(['Squirrel Hill South', 'Beechview', 'South Side Slopes', 'Allegheny Center', 'Highland Park', 'Sheraden', 'Beltzhoover', 'Crawford-Roberts', 'Hazelwood', 'East Liberty', 'Upper Lawrenceville', 'South Oakland', 'Elliott', 'Carrick', 'Mount Washington'])
# Display topValues as table
display(topValues)
```

Conclusion: 
Based on the results, we can conclude that Squirrel Hill South is the best neighborhood due to it having the most playgrounds.
